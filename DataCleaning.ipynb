```python
import pandas as pd
```

```python
df=pd.read_csv('df_all_before_cleaning.csv')
```

### Remove Duplicates

```python
duplicate = df[df.duplicated()]
```

```python
duplicate
```

```python
df.drop_duplicates(inplace = True)
```

### Convert data type 

```python
from dateutil import parser

df['start_time'] = df['start_time'].apply(lambda x:parser.parse(x))
df['end_time'] = df['end_time'].apply(lambda x:parser.parse(x))
```

### Clean 'member_type'

```python
df['member_type'].value_counts(dropna=False)
```

```python
df[df['member_type']=='Dependent'].shape
```

```python
## Considering 190 of 31,067,208 data points has member type as dependent
## These data points can be droped.

df.drop(df[df['member_type']=='Dependent'].index,inplace=True)
```

```python
df['member_type'].unique()
```

```python
## Now it is time to change 'member_type' to boolean type

member_type_dic={'member':'member',
            'Subscriber':'member',
            'casual':'non_member',
            'Customer':'non_member'}

df.replace({'member_type':member_type_dic},inplace=True)
```

```python
df.head()
```

```python
df['member_type'].unique()
```

```python
df['member_type'].value_counts(dropna=False)
```

### Clean 'trip_duration'

```python
df['trip_duration'].unique()
```

```python
df['trip_duration'].value_counts(dropna=False)
```

```python
df['total_seconds_cal']=(df['end_time']-df['start_time']).apply(lambda x:x.total_seconds())
```

```python
df['trip_duration']=df['total_seconds_cal']
```

```python
df['start_year']=df['start_time'].apply(lambda x:x.year)
```

```python
df['start_month']=df['start_time'].apply(lambda x:x.month)
```
